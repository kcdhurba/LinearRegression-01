```python
# Imports 

import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = {
    "x1": [1.1, 2.0, 3.2, 4.1, 5.0, 6.3, 7.5, 8.0, 9.2, 10.0],
    "x2": [2.0, 1.8, 3.0, 3.5, 4.2, 5.0, 6.0, 7.1, 8.3, 9.0],
    "x3": [3.0, 2.5, 4.2, 5.0, 6.0, 7.5, 8.0, 8.8, 10.0, 11.5],
    "x4": [1.5, 1.8, 2.0, 2.5, 3.0, 3.2, 3.8, 4.0, 4.5, 5.0],
    "x5": [2.5, 3.0, 4.1, 5.3, 6.8, 7.5, 8.3, 9.0, 10.2, 11.5],
    "y": [4.2, 5.8, 8.3, 10.9, 13.5, 16.8, 19.4, 21.2, 24.7, 27.3]
}

# Creating X_train as a NumPy array
X_train = np.array([data["x1"], data["x2"], data["x3"], data["x4"], data["x5"]]).T

# Target variable y
y_train = np.array(data["y"])
```

```python
# initializing random values of parameters W(weights) and b(bias) 

number_of_features = 5 
# Initialize weights with small random values
W_init = np.random.randn(number_of_features)

# Initialize bias with small random value 
b_init = np.random.randn()

print(f"Initial weight: {W_init}")
print(f"Initial bias: {b_init}")
```

```python
# Shapes of X_train, y_train 

print(X_train.shape), print(y_train.shape)
```

```python
# Lineplot of the data 

plt.figure(figsize=(10,6))
# Loop through features
for i in range(X_train.shape[1]):
    plt.scatter(X_train[:, i], y_train, label=f'Feature {i + 1}')

# Adding labels and title
plt.xlabel('Features')
plt.ylabel('Target (y)')
plt.title('Scatter Plot of Features vs Target')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# Prediction vector for single example

def predict(X, w, b ): 
    """
    single predict using LR
    Args: 
    X(ndarray): shape(n,) single example with multiple features 
    W(ndarray): shape(n,) model pararms 
    b(ndarray): model param
    
    Returns: 
    p(scalar): predictino
    """
    p = np.dot(X,w) + b 
    return p 
```

```python
x_ = X_train[0]
y_ = y_train[0]
predict(x_, W_init, b_init)
```

```python
# Cost with multiple variables 
def cost(X, y, W, b): 
    """
    compute cost args: 
    X(ndarray): (m,n) -> data with m examples with n features 
    y(ndarray): (m,) -> target values of length m 
    W(ndarray): (n,) -> model parameter
    b(scalar): model param

    Returns 
    cost(scalar): cost 
   
    """
    # number of examples
    m = X.shape[0] 
    # intial cost 
    cost = 0.0
    # loop through all the examples 
    for i in range(m): 
        f_wb_i = np.dot(X[i], W) + b
        cost += (f_wb_i - y[i]) **2 

    cost = cost/(2 * m)
    return cost 


```

```python
intial_cost = cost(X_train, y_train, W_init, b_init)
print(f'cost with intial parameters: {intial_cost}')
```
